```python
# plot the different discreet variables w.r.t the 'chocolate' class 
# we drop chocolate from the plotted columns 
#for var in discreet_vars[1:]:
#    fig=plt.figure()

```

```python
#  function to plot the distributions for each feature w.r.t the chocolate feature 
def plotPairs(data,feature1,feature2):
    figure, axes = plt.subplots(1,2)
    figure.suptitle(f"{feature1} - {feature2}")
    sns.set_style('whitegrid')
    sns.boxplot(x=feature1,y=feature2,ax=axes[0],data=data)
    sns.stripplot(x=feature1,y=feature2,ax=axes[0],data=data)
    sns.set_style('whitegrid')
    sns.violinplot(x=feature1,y=feature2,ax=axes[1],data=data)
    sns.stripplot(x=feature1,y=feature2,ax=axes[1],data=data)

# excude chocolate 
for var in discreet_vars[1:]:
    plotPairs(data,'chocolate',var)
```

To be able to draw some clear conclusions on the pairwise relations between chocolate and other features we can group for each binary value for each feature the number of observations that have chocolate.

```python
data['chocolate'].value_counts()
```

```python
# count for each feature the different feature values-observations 
print(data.groupby('bar').sum())
print(data.groupby('fruity').sum())
print(data.groupby('hard').sum())
print(data.groupby('nougat').sum())
print(data.groupby('chocolate').sum())
```

<u>From the above we can see for example that for the 'bar' feature 17 chocolate products (from the total chocolate products) are not 'bar' products but the other '20' chocolate products are 'bar' products .</u>

### 2. Feature Engineering

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import joblib
```

<ul>
<li> We split the dataset into train and test set and of course we use random_state and set a SEED so we can reproduce the model in production </li>
</ul>

```python
# We start by separating the set into train and test set 
# we drop the target variable from the x sets and also the 'competitorname' feature 
# which we have already dropped 
x_train,x_test,y_train,y_test = train_test_split(data.drop(["chocolate"],inplace=False,axis=1),data["chocolate"],test_size=0.2,random_state=0,stratify=data['chocolate'])
```

```python
print(x_train.shape)
print(x_test.shape) 
print(y_train.shape)
print(y_test.shape) 
```

```python
# we should also reset the index for the y_train, y_test 
y_train.reset_index(drop=True,inplace=True)
y_test.reset_index(drop=True,inplace=True)
```

Next we will explore whether the frequency of the classes (0 or 1) for chocolate is similar for both datasets (y_train and y_test)

```python
print ('\n |--- Y_TRAIN --- |  \n', y_train.value_counts(normalize=True))
print( '\n |--- Y_TEST --- |  \n', y_test.value_counts(normalize=True))
```

Indeed we can see that chocolate/non_chocolate frequencies of both y_train and y_test are similar so we don't have any strong bias in our test set.

5. Standardization

We need to standardize our features so the neural networks converge faster. We will use the MinMaxScaler to scale the values to the minimum and maximum values , for the continuous variables.
<u>Remember we fit the Scaler only on the TRAINING set and then transform both the training set and the test set</u>

```python
scaler = MinMaxScaler()

# fit scaler 
scaler.fit(x_train)


# transform X_test DataFrame
x_train = pd.DataFrame(
    scaler.transform(x_train),
    columns=x_train.columns
)

# transform X_test 
x_test= pd.DataFrame(
    scaler.transform(x_test),
    columns=x_train.columns
)
```

```python
# sanity check 
print(x_train.head(1))
print('\n')
print(x_test.head(1))
print('\n')
print(y_test.head(1))
print(y_train.head(1))
```
